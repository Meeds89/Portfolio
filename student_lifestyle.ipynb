# Relatório de Análise de Dados: Estilo de Vida e Desempenho Acadêmico

## 1. Introdução
Este relatório explora padrões de estilo de vida de estudantes e sua relação com o desempenho acadêmico (GPA). Analisamos dados sobre estudo, sono, atividades extracurriculares, atividades físicas e estresse.

### Objetivo
- Identificar fatores que impactam o GPA.
- Explorar como o estresse se relaciona com hábitos de vida.

---

## 2. Metodologia
### Dataset
- **Registros:** 2.000 estudantes.
- **Variáveis principais:** Horas de estudo, sono, atividades físicas, atividades extracurriculares, níveis de estresse e GPA.

### Etapas:
1. Carregamento e limpeza dos dados.
2. Análise exploratória.
3. Visualizações de correlação.

---

 #### Carregamento de Dados

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_lifestyle_dataset.csv")
```

---

#### Análise Exploratória

```python
# Distribuição de GPA
plt.hist(data=df, x="GPA", bins=7);
```

```python
# GPA médio e Desvio Padrão
print("Mean:", df["GPA"].mean().round(1))
print("Standard Deviation:", df["GPA"].std())
```

```python
# Quantidade de estudantes com nota maior ou igual a média
df[df["GPA"] >= df["GPA"].mean().round(1)]["Student_ID"].count()
```

---

## 3. Resultados
### Relação entre Estudo e GPA
- **Insight:** Estudantes que estudam mais têm maior GPA.
- **Observação:** Estresse aumenta com horas de estudo.

```python
# Relação "Estudos X GPA"
sns.lmplot(data=df, x="Study_Hours_Per_Day", y="GPA", line_kws={'color': 'red'})
plt.xlabel("Study Hours Per Day")
plt.title("Study Hours Per Day X GPA");
```

```python
# Relação "Estudos X Estresse"
sns.barplot(data=df, y="Study_Hours_Per_Day", x="Stress_Level")
plt.xlabel("Stress Level")
plt.ylabel("Study Hours Per Day")
plt.title("Study Hours Per Day X GPA");
```

### Impacto do Sono
- **Insight:** Dormir menos de 7 horas/dia eleva o estresse, mas não afeta diretamente o GPA.

```python
# Relação "Sono X GPA"
sns.lmplot(data=df, x="Sleep_Hours_Per_Day", y="GPA", line_kws={'color': 'red'})
plt.xlabel("Sleep Hours Per Day")
plt.title("Sleep Hours Per Day X GPA");
```

```python
# Relação "Sono X Estresse"
sns.barplot(data=df, y="Sleep_Hours_Per_Day", x="Stress_Level");
```

### Atividades Extracurriculares
- **Insight:** Não há impacto significativo no GPA.

```python
# Relação "Atividades Extras X GPA"
sns.lmplot(data=df, x="Extracurricular_Hours_Per_Day", y="GPA", line_kws={'color': 'red'})
plt.xlabel("Extracurricular Hours Per Day X GPA")
plt.title("Extracurricular Hours Per Day X GPA");
```

### Atividades Físicas
- **Insight:** Reduzem o estresse, mas muitas horas podem reduzir o GPA.

```python
# Relação "Atividades Físicas X GPA"
sns.lmplot(data=df, x="Physical_Activity_Hours_Per_Day", y="GPA", line_kws={'color': 'red'})
plt.xlabel("Daily Physical Activity")
plt.title("Daily Physical Activity X GPA");
```

```python
# Relação "Atividades Físicas X Estresse"
sns.barplot(data=df, y="Physical_Activity_Hours_Per_Day", x="Stress_Level")
plt.xlabel("Stress Level")
plt.ylabel("Daily Physical Activity")
plt.title("Daily Physical Activity X Stress Level");
```

---

## 5. Conclusão e Recomendações
- **Conclusão:** O equilíbrio entre estudo, sono e atividades físicas é crucial.
- **Recomendações:**
  1. Priorizar 7-9 horas de sono.
  2. Balancear estudo com atividades físicas para controlar o estresse.
  3. Reduzir o foco em atividades extracurriculares, pois o impacto no GPA é limitado.
